2️⃣ Qué es realmente la práctica (PPT)

Es una simulación real de startup.

Se exige:
	•	Trabajo con SCRUM
	•	Backlog en Trello
	•	Plan de negocio inicial
	•	MVP funcional
	•	No hace falta completar todo el backlog
	•	Se valora el avance real y la calidad

3️⃣ Pipeline obligatorio del proyecto

Debéis cubrir:
	1.	Definición del problema
	2.	Recolección de datos
	3.	Preprocesamiento
	4.	EDA
	5.	Modelado
	6.	Evaluación
	7.	Optimización
	8.	Despliegue
	9.	Monitoreo
	10.	Documentación

No es opcional. Es la estructura base  ￼.

Resumen y Recomendaciones del Profesor (transcripción masterclass TFM)


1) Qué quiere realmente que entreguéis
	•	Un proyecto “de principio a fin” como si fuera un producto real: problema → datos → modelo(s) → demo → (mínimo) despliegue/plan de despliegue → documentación.
	•	Un MVP (no producción). Repite que no se espera 100%; se espera máximo avance posible en 2 sprints.
	•	Trabajo en equipo (recomendado fuerte). Individual se permite, pero “no le gustaría”.

2) Regla nº1 antes de tocar código

Antes de escribir una sola línea:
	•	Responder en un párrafo:
	1.	Qué problema resolvemos
	2.	Para quién
	3.	Cómo sabremos que lo resolvimos bien
Si no podéis responder eso, no empecéis.

3) Mentalidad “ingeniero”, no “estudiante”
	•	Considerar escalabilidad y costes desde el inicio.
	•	“El modelo más grande no siempre es mejor”: un modelo pequeño bien ajustado puede ganar a uno grande mal montado.
	•	Se valora justificación de decisiones (por qué hiciste X y no Y).

4) SCRUM y trabajo realista
	•	2 sprints / ~3 semanas.
	•	Usar backlog de historias de usuario (de lo primero a hacer).
	•	Rotación de roles: todos deben tocar (o al menos entender) datos, modelado, despliegue, documentación.
	•	Coordinación simple aceptada (Trello + WhatsApp/Discord/Telegram). No exige “Scrum perfecto”.

5) Datos: cómo resolver el “bloqueo típico”
	•	Datos pueden ser: internos, públicos, scraping, Hugging Face, sintéticos.
	•	Si no hay datos: crear/usar sintéticos (reconoce que en imágenes es más difícil).
	•	Mencionar cumplimiento normativo (ej.: GDPR) y explicar limitaciones si aplica.

6) Modelado y evaluación: no vale “porque sí”
	•	Elegir modelo y explicar por qué.
	•	Evaluación: admite que a veces es difícil, pero exige alguna forma razonable:
	•	métricas clásicas si aplica,
	•	o evaluación simple (p.ej., set de prompts y revisión sistemática) si es LLM.
	•	Iterar: hiperparámetros/ajuste, aunque sea básico.

7) Despliegue: “con pinzas”, pero defendible
	•	No exige infraestructura “pro”.
	•	Vale demo en Colab / local si es estable para la demo.
	•	Muy importante: llevar una slide preparada de “¿cuánto costaría al mes?” / necesidades de hardware y qué implicaría producción.
	•	Justificar por qué elegiste ese tipo de despliegue.

8) Documentación: crítica y penalizable si falta
	•	Si entregáis código sin documentación, se complica evaluar y puede penalizar.
	•	Recomienda documentar en paralelo, no al final.
	•	Debe permitir reproducibilidad (que ellos lo puedan correr).
	•	Contenido mínimo recomendado:
	•	problema + suposiciones,
	•	métricas y justificación,
	•	métodos,
	•	resultados,
	•	sesgos/limitaciones,
	•	lecciones aprendidas + siguientes pasos.

9) Presentación: lo que más enfatiza (y lo que penaliza)
	•	Presentación: 10–15 min. Pasarse de 15 min penaliza mucho.
	•	Tono: hablar como si delante hubiera comité ejecutivo / VC, no técnicos.
	•	En esos 10–15 min NO entrar en detalle tipo “finetuning, hiperparámetros, etc.”.
	•	Estructura sugerida:
	•	problema,
	•	enfoque/solución (alto nivel),
	•	cómo lo hicisteis (alto nivel),
	•	implicaciones para producción/costes,
	•	demo,
	•	aprendizajes + conclusiones.
	•	Recomendación práctica: hacer un vídeo de la demo (para evitar fallos en directo).
	•	Presentan todos, repartido tipo “obra de teatro” (cada uno se sabe su parte).
	•	Luego, en Q&A sí entran a técnica y revisan código.

10) Reusar prácticas previas
	•	Sí se puede partir de una práctica (ej. la de LLMs) pero no entregar lo mismo: hay que expandir (más datos, más pipeline, más modelado, más robustez, mejor demo, etc.).

⸻

Recomendaciones operativas (derivadas de lo que repite)
	1.	Definir problema (3 preguntas) + criterio de éxito en 1 párrafo.
	2.	Montar Trello con historias de usuario y 2 sprints.
	3.	Asegurar demo estable (ideal vídeo).
	4.	Preparar 1 slide de costes/producción (aunque no lo despleguéis “pro”).
	5.	Documentar desde el día 1 y dejar instrucciones para reproducir.
	6.	Ensayar para clavar 10–15 min sin pasarse.
